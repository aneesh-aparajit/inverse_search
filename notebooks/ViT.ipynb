## Patch

$$
\mathbb{R}^{H\times W\times C} \rightarrow \mathbb{R}^{N\times (P^2\cdot C)}
$$

- P - Patch Size

```python
P = 16
H = 224
W = 224
C = 3
N = int((H*W)/P**2)
print(P, H, W, C, N)
```

```python
img_new = copy.deepcopy(img)
img_new.shape
```

```python
img_patches = img_new.reshape(N, P*P*C)
img_patches.shape
```

```python
plt.figure(figsize=(P, P))
plt.imshow(img_patches[:, :]);
```
